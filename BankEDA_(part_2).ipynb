## Analyze Loyalty Classification and Income Band

### Subtask:
Analyze the relationship between Loyalty Classification and Income Band by looking at the count of customers in each segment and their average financial metrics like Estimated Income, Bank Deposits, and Bank Loans. Visualize these relationships using appropriate plots (e.g., stacked bar chart for counts, grouped bar charts for averages).


**Reasoning**:
First, I'll calculate the number of customers in each combination of 'Loyalty Classification' and 'Income Band' using `groupby()` and `size()`, then `unstack()` to prepare for plotting. After that, I'll calculate the average of `Estimated Income`, `Bank Deposits`, and `Bank Loans` for each of these combinations. This will fulfill steps 1-4 of the instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Calculate the number of customers in each combination of 'Loyalty Classification' and 'Income Band'
loyalty_income_counts = df.groupby(['Loyalty Classification', 'Income Band']).size().unstack(fill_value=0)

print("Customer Counts by Loyalty Classification and Income Band:")
display(loyalty_income_counts)

# 2-4. Calculate the average financial metrics for each combination
financial_metrics = ['Estimated Income', 'Bank Deposits', 'Bank Loans']
average_metrics = df.groupby(['Loyalty Classification', 'Income Band'])[financial_metrics].mean()

print("\nAverage Financial Metrics by Loyalty Classification and Income Band:")
display(average_metrics)
```

**Reasoning**:
The previous steps successfully calculated the customer counts and average financial metrics per loyalty classification and income band. Now, I will visualize these results as per the instructions: a stacked bar chart for customer counts and grouped bar charts for the average financial metrics. This will complete the visualization requirements of the subtask.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 5. Create a stacked bar chart for customer counts
loyalty_income_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Customer Counts by Loyalty Classification and Income Band')
plt.xlabel('Loyalty Classification')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Income Band')
plt.tight_layout()
plt.show()

# 6. Create grouped bar charts for average financial metrics
for metric in financial_metrics:
    plt.figure(figsize=(12, 7))
    sns.barplot(x='Loyalty Classification', y=metric, hue='Income Band', data=average_metrics.reset_index())
    plt.title(f'Average {metric} by Loyalty Classification and Income Band')
    plt.xlabel('Loyalty Classification')
    plt.ylabel(f'Average {metric}')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Income Band')
    plt.tight_layout()
    plt.show()
```

## Analyze Age-based Financial Behavior across Income Bands

### Subtask:
Investigate how financial behaviors (e.g., Bank Deposits, Bank Loans, Credit Card Balance) vary with Age across different Income Band categories. Create plots (e.g., line plots of mean values or scatter plots) to visualize these trends, providing insights into different life stages and income levels.


**Reasoning**:
First, I'll define the financial metrics to analyze. Then, I'll group the DataFrame by 'Age' and 'Income Band' and calculate the mean of these metrics. Finally, I'll iterate through each financial metric to create a line plot showing its variation with age across different income bands, as instructed.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define a list of financial metrics to analyze
financial_metrics_age = ['Bank Deposits', 'Bank Loans', 'Credit Card Balance']

# 2. Group the DataFrame by 'Age' and 'Income Band', then calculate the mean of the selected financial metrics
age_income_metrics = df.groupby(['Age', 'Income Band'])[financial_metrics_age].mean().reset_index()

print("Average Financial Metrics by Age and Income Band (first 5 rows):")
display(age_income_metrics.head())

# 3. For each financial metric, create a line plot
for metric in financial_metrics_age:
    plt.figure(figsize=(12, 7))
    sns.lineplot(data=age_income_metrics, x='Age', y=metric, hue='Income Band', marker='o')
    plt.title(f'Average {metric} by Age and Income Band')
    plt.xlabel('Age')
    plt.ylabel(f'Average {metric}')
    plt.legend(title='Income Band')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
The previous code block executed successfully and generated the requested plots, but it produced a `FutureWarning` from pandas. To address this warning and ensure forward compatibility, I will add `observed=False` to the `groupby` call, which is the current default behavior that will change in a future pandas version.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define a list of financial metrics to analyze
financial_metrics_age = ['Bank Deposits', 'Bank Loans', 'Credit Card Balance']

# 2. Group the DataFrame by 'Age' and 'Income Band', then calculate the mean of the selected financial metrics
age_income_metrics = df.groupby(['Age', 'Income Band'], observed=False)[financial_metrics_age].mean().reset_index()

print("Average Financial Metrics by Age and Income Band (first 5 rows):")
display(age_income_metrics.head())

# 3. For each financial metric, create a line plot
for metric in financial_metrics_age:
    plt.figure(figsize=(12, 7))
    sns.lineplot(data=age_income_metrics, x='Age', y=metric, hue='Income Band', marker='o')
    plt.title(f'Average {metric} by Age and Income Band')
    plt.xlabel('Age')
    plt.ylabel(f'Average {metric}')
    plt.legend(title='Income Band')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

## Analyze Risk Weighting and Financial Metrics

### Subtask:
Examine how Risk Weighting (which seems to be an integer from 1 to 4 based on the df.describe() output) relates to various financial metrics such as Bank Loans, Credit Card Balance, and Estimated Income. Use bar plots to show average values for each risk category.


**Reasoning**:
I will define the financial metrics of interest, then group the DataFrame by 'Risk Weighting' to calculate the average for these metrics, and finally generate bar plots for each metric to visualize their relationship with 'Risk Weighting', as instructed.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define a list of financial metrics to analyze
financial_metrics_risk = ['Bank Loans', 'Credit Card Balance', 'Estimated Income']

# 2. Group the DataFrame df by the 'Risk Weighting' column and calculate the mean of the selected financial metrics
risk_metrics_avg = df.groupby('Risk Weighting')[financial_metrics_risk].mean().reset_index()

print("Average Financial Metrics by Risk Weighting:")
display(risk_metrics_avg)

# 3-7. For each financial metric, create a bar plot
for metric in financial_metrics_risk:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Risk Weighting', y=metric, data=risk_metrics_avg, palette='viridis')
    plt.title(f'Average {metric} by Risk Weighting')
    plt.xlabel('Risk Weighting')
    plt.ylabel(f'Average {metric}')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
The previous code block produced a `FutureWarning` related to `sns.barplot` and the `palette` argument. To resolve this, I will modify the `sns.barplot` call to explicitly set `hue='Risk Weighting'` and `legend=False`, as recommended by the warning.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define a list of financial metrics to analyze
financial_metrics_risk = ['Bank Loans', 'Credit Card Balance', 'Estimated Income']

# 2. Group the DataFrame df by the 'Risk Weighting' column and calculate the mean of the selected financial metrics
risk_metrics_avg = df.groupby('Risk Weighting')[financial_metrics_risk].mean().reset_index()

print("Average Financial Metrics by Risk Weighting:")
display(risk_metrics_avg)

# 3-7. For each financial metric, create a bar plot
for metric in financial_metrics_risk:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Risk Weighting', y=metric, hue='Risk Weighting', data=risk_metrics_avg, palette='viridis', legend=False)
    plt.title(f'Average {metric} by Risk Weighting')
    plt.xlabel('Risk Weighting')
    plt.ylabel(f'Average {metric}')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
